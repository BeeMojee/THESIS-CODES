```python
df = pd.read_csv('/content/drive/My Drive/cleaned_total_reviews.csv')
df.head()
```

```python
# Grouping the data by bank and manual_labels to count the occurrences
bank_sentiment_counts = df.groupby(['Bank', 'manual_labels']).size().unstack(fill_value=0)
```

```python
import matplotlib.pyplot as plt
```

```python
# Plotting the bar chart
plt.figure(figsize=(8, 5))
bank_sentiment_counts.plot(kind='bar', stacked=False, color=['red', 'blue', 'green'], alpha=0.7, ax=plt.gca())
plt.title('Review Sentiment per Bank')
plt.xlabel('Bank')
plt.ylabel('Number of Reviews')
plt.legend(title='Sentiment')
plt.grid(False)
plt.show()
```
