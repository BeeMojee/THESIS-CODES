```python
df = pd.read_csv('/content/drive/My Drive/cleaned_total_reviews.csv')
df.head()
```

```python
# Filter data for negative manual labels
negative_reviews =df[df['manual_labels'].str.lower() == 'negative']
```

```python
nltk.download('punkt')
```

```python
nltk.download('stopwords')
```

```python
# Get the set of English stopwords using a different variable name
english_stopwords = stopwords.words('english')
```

```python
stop_words = set(english_stopwords)
```

```python
# Function to process reviews and generate a word cloud
def generate_word_cloud(bank_data, bank_name):
    all_reviews = []  # Initialize an empty list for all reviews

    for review in bank_data['review']:
        words = word_tokenize(review.lower())  # Convert to lower case and tokenize
        filtered_words = [word for word in words if word.isalpha() and word not in stop_words]
        all_reviews.extend(filtered_words)  # Extend the list of all reviews

    text = ' '.join(all_reviews)  # Join all words to create a single string

    # Create and display the word cloud
    wordcloud = WordCloud(width=800, height=400, background_color='black').generate(text)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.title(f"Word Cloud for {bank_name}")
    plt.show()
```

```python
# Generate word clouds for each bank for negative reviews
banks = df['Bank'].unique()
for bank in banks:
    bank_data = negative_reviews[negative_reviews['Bank'] == bank]
    if not bank_data.empty:
        generate_word_cloud(bank_data, bank)
    else:
        print(f"No negative reviews for {bank}.")
```
