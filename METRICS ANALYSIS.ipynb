```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
#RATINGS LABEL
df = pd.read_csv('/content/drive/My Drive/RATINGS_LABEL_TF-IDF.csv')
df.head()
```

```python
# Melting the DataFrame
melted_data = pd.melt(df, id_vars=['Models', 'Labels'], value_vars=['Precision', 'Recall', 'F1_score'],
                      var_name='Metric', value_name='Value')
```

```python
# Creating a combined column for 'Metric' and 'Labels'
melted_data['Metric_Label'] = melted_data['Metric'] + ' - ' + melted_data['Labels']
```

```python
# Plotting the data with the new combined hue
plt.figure(figsize=(18, 10))
sns.barplot(data=melted_data, x='Models', y='Value', hue='Metric_Label', ci=None, dodge=True)
# Adding details to the plot
plt.title('Model-TF_IDF Performance Across Metrics and Labels (Ratings)')
plt.ylabel('Metric Value (%)')
plt.xlabel('Models')
plt.legend(title='Metric and Label', bbox_to_anchor=(1.05, 1), loc='upper left')
# Show the plot
plt.show()
```
